# Sentence Pair Classification Task
In this notebook, I will be working on the sentence pair classification task explained in *Glyce: Glyph-vectors for Chinese Character Representations* from this [paper](https://arxiv.org/pdf/1901.10125.pdf). There are four datasets that were used for this task:


*   BQ (binary classification task [Bowman])
*   LCQMC (binary classification task [Liu])
*   XNLI (three class classifcation task [William and Bowman])
*   NLPCC-DBQA (binary classification task [github](https://github.com/xxx0624/QA_Model://)

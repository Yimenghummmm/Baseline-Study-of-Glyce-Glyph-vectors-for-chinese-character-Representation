```
import pandas as pd
import csv

bq_train_TSV = './dataset/sent_pair/bq/train.tsv'
bq_test_TSV  = './dataset/sent_pair/bq/test.tsv'
```

```
def read_tsv(input_file):
  with open(input_file, "r") as f:
    reader = csv.reader(f, delimiter="\t")
    lines = []
    for line in reader:
      lines.append(line)
    return lines
```

```
bq_train = read_tsv(bq_train_TSV)
```
